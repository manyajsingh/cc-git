````python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
iris=sns.load_dataset('iris')
x=iris['petal_length']
y=iris['petal_width']
plt.scatter(x,y)
plt.xlabel('petal length')
plt.ylabel('petal width')
````

````python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.4, random_state=25)
x_train=np.array(x_train).reshape(-1,1)
x_test=np.array(x_test).reshape(-1,1)
````

````python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
c=lr.intercept_
m=lr.coef_
y_pred_train=m*x_train + c
y_pred_train.flatten()
````

````python
y_pred_train1=lr.predict(x_train)
y_pred_train1
````

````python
plt.scatter(x_train, y_train)
plt.plot(x_train, y_pred_train1, color='red')
plt.xlabel('petal length')
plt.ylabel('petal width')
````

````python
y_pred_test1=lr.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred_test1, color='red')
plt.xlabel('petal length')
plt.ylabel('petal width')
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
x= np.random.randint(0,20,100)
noise=np.random.randn(100)*2
y= 3*x + 1 + noise
print(x.shape, y.shape)
plt.scatter(x,y, cmap='magma')
plt.show()
````

````python
print(x)
````

````python
x=x.reshape(100,1)
print(x)
````

````python
def hypothesis(x, weights):
  y_hat=weights[0] + weights[1]*x # x: real input value (time spent)
                                  # weights: list of w0 and w
  return y_hat
````

````python
def error(x,y,weights):
  n=x.shape[0]                    # x: input n data points
                                  # y: target n data points
                                  # weights: list of w0 and w1
  total_error=0.0
  for i in range(n):
    y_hat_i= hypothesis(x[i], weights)
    total_error+=(y[i] -y_hat_i)**2
  return total_error/n
````

````python
def gradient(x,y,weights):
  n=x.shape[0]
  grad=np.zeros(2)
  for i in range(n):
    y_hat_i= hypothesis(x[i], weights)
    grad[0] += (y_hat_i - y[i])
    grad[1] += ((y_hat_i -y[[i]])*x[i])
  return 2*grad/n
````

````python
def gradient_descent(x,y,max_itr=50, learning_rate=0.001):
  weights= np.random.randn(2)
  error_list=[]
  for i in range(max_itr):
    e=error(x,y,weights)
    error_list.append(e)
    grad=gradient(x,y,weights)
    weights[0]-=learning_rate*grad[0]
    weights[1]-=learning_rate*grad[1]
  return weights, error_list
````

````python
opt_weights, error_list= gradient_descent(x,y)
opt_weights.round()
````

````python
plt.plot(error_list)
````

````python
print(hypothesis(x[0], opt_weights))
print(y[0])
for i in [0,19,21]:
  print(opt_weights[1]*i + opt_weights[0])
````

````python
x1,y1= 0, 0.69
x2,y2=19, 58.21
plt.scatter(x,y)
plt.plot([x1,x2], [y1,y2], color='red')
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv('insurance.csv')
````

````python
df.isnull().sum()
````

````python
r2_train=r2_score(y_train, y_pred_train)
r2_test=r2_score(y_test, y_pred_test)
print(r2_train, r2_test)
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Heart.csv')
df.head()
````

````python
df.drop(columns='Unnamed: 0', axis=1, inplace=True)
````

````python
df['ChestPain']=df['ChestPain'].astype('category')
df['ChestPain']=df['ChestPain'].cat.codes

df['AHD']=df['AHD'].astype('category')
df['AHD']=df['AHD'].cat.codes

df['Thal']=df['Thal'].astype('category')
df['Thal']=df['Thal'].cat.codes
df.head()
````

````python
df.isnull().sum()
````

````python
# to fill null values with random inputs:
# df['Ca']=df['Ca'].fillna(value=2.0)
# df['Thal']=df['Thal'].fillna(value=1)

# to drop the rows having null values
df.dropna(inplace=True)
````

````python
x=df.drop(columns='AHD')
y=df['AHD']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state=21)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=0).fit(x_train_scaled, y_train)

print(lr.predict(x_train_scaled)) # predicted values of AHD (our dependant variable-y axis) (trained)
print(lr.score(x_train_scaled, y_train)) # 88% data is accurate (trained)

print(lr.predict(x_test_scaled)) # predicted values of AHD (our dependant variable-y axis) (tested)
print(lr.score(x_test_scaled, y_test)) # 80% data is accurate (tested)
````

````python
# to improve the  model:
lr1=LogisticRegression(random_state=0, C=1, fit_intercept=True).fit(x_train_scaled, y_train)
# C- a regulization parameter that minimizes the extreme values
# fit_intercept for accurate reading
print(lr1.predict(x_train_scaled))
print(lr1.score(x_train_scaled, y_train))

print(lr1.predict(x_test_scaled))
print(lr1.score(x_test_scaled, y_test))
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
train=pd.read_csv('fashion-mnist_train.csv')
test=pd.read_csv('fashion-mnist_test.csv')
````

````python
x_train=train.drop(columns='label')
y_train=train['label']

x_test=test.drop(columns='label')
y_test=test['label']

x_train_list=x_train.values.tolist() # to convert dataframe to list (2D)
x_test_list=x_test.values.tolist()
````

````python
plt.imshow(np.reshape(x_train_list[0], (28,28,1)))
````

````python
train.dropna(inplace=True)
test.dropna(inplace=True)

x_train.dropna(inplace=True)
x_test.dropna(inplace=True)

y_train.dropna(inplace=True)
y_test.dropna(inplace=True)
````

````python
y_train.drop(21770, inplace=True, axis=0)
````

````python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=12).fit(x_train, y_train)
````

````python
print(lr.score(x_train, y_train))
print(lr.score(x_test, y_test))
````

````python
lr.predict([x_train_list[3]]) # correct prediction
````

````python
from sklearn import metrics
y_pred_test=lr.predict(x_test)
cm=metrics.confusion_matrix(y_test, y_pred_test)

plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt='0.3f', linewidth=0.5, square=True, cbar=True)
plt.ylabel('Actual values')
plt.xlabel('Predicted values')
plt.show()
````

````python
# precision = TP/(TP+FP)
# precision for label 9: 941/(941+46+41+2)= 0.91

# recall = TP/(TP+FN)
# recall for label 9: 941/(941+13+43+3)= 0.941

# F1 score= 2*((precision*recall)/(precision+recall))
# F1 score for label 9: 2*0.91*0.941/(0.91+0.941)= 0.925

print(metrics.classification_report(y_test, y_pred_test))
````
